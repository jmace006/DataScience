# Line Charts Project


```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('data/percent-bachelors-degrees-women-usa.csv')

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (171/255, 171/255, 171/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

first_row = 0
last_row = 5

fig = plt.figure(figsize=(16, 20))

# In the first column:
# Generate line charts for both male and female percentages for every degree in stem_cats.
# Add text annotations for Women and Men in the topmost and bottommost plots.

for sp in range(1,18,3):
    
    current_row = int(sp/3)
    
    # Generate a 6 row by 3 column grid of subplots.
    ax = fig.add_subplot(6,3,sp)
  
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[current_row]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[current_row]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_grey, alpha=0.3)
    ax.set_title(stem_cats[current_row])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    if current_row == first_row:
        ax.text(2005, 80, 'Men')
        ax.text(2004, 15, 'Women')
    elif current_row == last_row:
        ax.text(2005, 75, 'Men')
        ax.text(2004, 22, 'Women')
        ax.tick_params(labelbottom='on')

# In the second column:
# Generate line charts for both male and female percentages for every degree in lib_arts_cats.
# Add text annotations for Women and Men for only the topmost plot (since the lines overlap at the end in the bottommost plot).

for sp in range(2,17,3):
    
    current_row = int(sp/3)
    
    # Generate a 6 row by 3 column grid of subplots.
    
    ax = fig.add_subplot(6,3,sp)
  
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[current_row]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[current_row]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_grey, alpha=0.3)
    ax.set_title(lib_arts_cats[current_row])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    if current_row == first_row:
        ax.text(2005, 75, 'Men')
        ax.text(2004, 22, 'Women')
    elif current_row == last_row - 1:
        ax.tick_params(labelbottom='on')


# In the third column:
# Generate line charts for both male and female percentages for every degree in other_cats.
# Add text annotations for Women and Men in the topmost and bottommost plots.

for sp in range(3,19,3):
  
    current_row = int(sp/3)-1
   
    # Generate a 6 row by 3 column grid of subplots.
    
    ax = fig.add_subplot(6,3,sp)
  
    ax.plot(women_degrees['Year'], women_degrees[other_cats[current_row]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[current_row]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_grey, alpha=0.3)
    
    ax.set_title(other_cats[current_row])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    if current_row == first_row:
        ax.text(2005, 90, 'Men')
        ax.text(2003, 5, 'Women')
    elif current_row == last_row:
        ax.text(2005, 62, 'Men')
        ax.text(2004, 35, 'Women') 
        ax.tick_params(labelbottom='on')
        
plt.savefig("gender_degrees.png")
plt.show()
```
